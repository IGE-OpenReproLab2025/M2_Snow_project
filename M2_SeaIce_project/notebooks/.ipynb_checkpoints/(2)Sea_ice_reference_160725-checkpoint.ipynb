```python
### Je fais un test

files = os.listdir("/home/jovyan/M2_SeaIce_project/Data/input_data/ref/")

nc_files = [filename for filename in files if filename.endswith(".nc")]
nc_files
```

```python
seaice_cover_dict = {}

for i, files in enumerate(nc_files):
    model_name = files.split("_")[0]
    
    path_sicbcs = os.path.join("/home/jovyan/M2_SeaIce_project/Data/input_data/ref/", files)
    path_areacella = os.path.join("/home/jovyan/M2_SeaIce_project/Data/input_data/areacella/", files.replace("sea_ice", "gridarea_seaice"))
    path_landfrac = os.path.join("/home/jovyan/M2_SeaIce_project/Data/input_data/landmask/", files.replace("sea_ice_ref", "Landmask"))
                                 
    # use of functions defined in the module
    ds_sicbcs, ds_area, ds_landfrac = get_data_ref(path_sicbcs, path_areacella, path_landfrac)
    sicbcs, seaice_cover, monthly_mean = extraction_variable_and_monthly_mean_ref(ds_sicbcs, landfrac=ds_landfrac["landfrac"])
    areacella = ds_area["cell_area"]
    seaice_cover_km2 = seaice_surface_calculation(seaice_cover, areacella)
    seaice_cover_dict[model_name] = seaice_cover_km2

    plot_seaice_cover_basic_nh_ref(sicbcs, title=files[:15])
    
    ds_sicbcs.close()
    ds_area.close()
```
