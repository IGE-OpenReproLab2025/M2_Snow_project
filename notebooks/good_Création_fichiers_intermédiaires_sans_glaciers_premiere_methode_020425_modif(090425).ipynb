## Test de filtre pour ignorer les glaciers

```python
sftgif = sftgif_BCC_CSM2["sftgif"]
```

```python
print("snow_cover_ini lat:", snow_cover_ini.coords['lat'])
print("snow_cover_ini lon:", snow_cover_ini.coords['lon'])

print("sftgif lat:", sftgif.coords['lat'])
print("sftgif lon:", sftgif.coords['lon'])
```

```python
# arrondi à 4 décimales pour éviter légères diff
snow_cover_ini.coords['lat'] = snow_cover_ini.coords['lat'].round(4)
snow_cover_ini.coords['lon'] = snow_cover_ini.coords['lon'].round(4)

sftgif.coords['lat'] = sftgif.coords['lat'].round(4)
sftgif.coords['lon'] = sftgif.coords['lon'].round(4)

# alignement des deux ds
snow_cover_ini_aligned = snow_cover_ini.reindex_like(sftgif, method='nearest')

# applique le masque
glacier_mask = sftgif >= 100
snow_cover_masked = snow_cover_ini_aligned.where(~glacier_mask, other=np.nan)
```

```python
#snow_cover_masked.to_netcdf('snow_cover_masked_BCC_CSM2.nc')
```

```python
glacier_mask.plot()
```

```python
snow_cover_masked_1800 = snow_cover_masked.isel(time=1850)
snow_cover_masked_1800

plt.figure(figsize=(10, 5))
snow_cover_masked_1800.plot()
```

## Calcul expansion neige avec areacella

```python
snow_cover_masked = snow_cover_masked.sel(time=slice("1995", "2014"))
```

```python
snow_cover_masked = snow_cover_masked.groupby("time.month").mean(dim="time")
```

```python
march_snow_cover = snow_cover_masked.sel(month=3)

plt.figure(figsize=(10, 5))
march_snow_cover.plot()
```

```python
monthly_mean = snow_cover_masked.mean(dim=("lat", "lon"))

plt.plot(range(1, 13), monthly_mean, marker="o")
plt.xlabel("Mois")
plt.ylabel("Couverture neigeuse moyenne")
plt.title("Moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

```python
path_areacella_BCC_CSM2 = os.path.join("~/private-storage/masques/masques/pseudo_areacella/", "areacella_BCC-CSM2-MR.nc")

areacella_BCC_CSM2 = xr.open_dataset(path_areacella_BCC_CSM2)
areacella_BCC_CSM2
```

```python
cell_area_ini = areacella_BCC_CSM2["areacella"]

cell_area_ini.plot()
```

```python
janvier = snow_cover_masked.isel(month=0)
janvier.plot()
```

```python
print("snow_cover_masked lat:", snow_cover_masked.coords['lat'])
print("snow_cover_masked lon:", snow_cover_masked.coords['lon'])

print("cell_area_ini lat:", cell_area_ini.coords['lat'])
print("cell_area_ini lon:", cell_area_ini.coords['lon'])
```

```python
snow_cover_masked.coords['lat'] = snow_cover_masked.coords['lat'].round(4)
snow_cover_masked.coords['lon'] = snow_cover_masked.coords['lon'].round(4)

cell_area_ini.coords['lat'] = cell_area_ini.coords['lat'].round(4)
cell_area_ini.coords['lon'] = cell_area_ini.coords['lon'].round(4)

snow_cover_masked_aligned = snow_cover_masked.reindex_like(cell_area_ini, method='nearest')
```

```python
snow_cover_km2 = (snow_cover_masked_aligned * (cell_area_ini * 1e-6) ).groupby(snow_cover_masked_aligned["month"]).sum(dim=["lat", "lon"], skipna=True)

print(snow_cover_km2)
print(snow_cover_km2.shape)
```

```python
plt.plot(range(1, 13), snow_cover_km2, marker="o")
plt.xlabel("Mois")
plt.ylabel("Superficie enneigée (km²)")
plt.title("Superficie moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

## test pour le même modèle after regridding

```python
path_BCC_CSM2_reprojete = os.path.join("~/private-storage/interp_models_hist_bis/", "BCC-CSM2-MR_hist_reprojete.nc")

BCC_CSM2_reprojete = xr.open_dataset(path_BCC_CSM2_reprojete)
BCC_CSM2_reprojete
```

```python
snow_cover_ini_reprojete = BCC_CSM2_reprojete["snc"]
snow_cover_ini_reprojete
```

### masque associé

```python
path_interp_sftgif_BCC_CSM2 = os.path.join("~/private-storage/interp_sftgif_bis", "BCC-CSM2-MR_sftgif_reprojete.nc")

interp_sftgif_BCC_CSM2 = xr.open_dataset(path_interp_sftgif_BCC_CSM2)
interp_sftgif_BCC_CSM2
```

### filtre pour ignorer les glaciers

```python
sftgif_reprojete = interp_sftgif_BCC_CSM2["sftgif"]
```

```python
# applique le masque
glacier_mask_repro = sftgif_reprojete > 50
snow_cover_masked_repro = snow_cover_ini_reprojete.where(~glacier_mask_repro, other=np.nan)
```

```python
snow_cover_masked_repro.to_netcdf('snow_cover_masked_repro_BCC_CSM2.nc')
```

```python
snow_cover_masked_1900 = snow_cover_masked_repro.isel(time=1850)
snow_cover_masked_1900

plt.figure(figsize=(10, 5))
snow_cover_masked_1900.plot()
```

### calcul expansion neige avec areacella

```python
snow_cover_masked_repro = snow_cover_masked_repro.sel(time=slice("1995", "2014"))
```

```python
snow_cover_masked_repro = snow_cover_masked_repro.groupby("time.month").mean(dim="time")
```

```python
monthly_mean_repro = snow_cover_masked_repro.mean(dim=("lat", "lon"))

plt.plot(range(1, 13), monthly_mean_repro, marker="o")
plt.xlabel("Mois")
plt.ylabel("Couverture neigeuse moyenne")
plt.title("Moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

```python
path_areacella_BCC_CSM2_recalcule = os.path.join("~/private-storage/areacella_recalcule_cdo_bis", "BCC-CSM2-MR_gridarea_hist_recalcule.nc")

areacella_BCC_CSM2_recalcule = xr.open_dataset(path_areacella_BCC_CSM2_recalcule)
areacella_BCC_CSM2_recalcule
```

```python
cell_area_ini_recalcule = areacella_BCC_CSM2_recalcule["cell_area"]
```

```python
snow_cover_km2_recalcule = (snow_cover_masked_repro * (cell_area_ini_recalcule * 1e-6) ).groupby(snow_cover_masked_repro["month"]).sum(dim=["lat", "lon"], skipna=True)

print(snow_cover_km2_recalcule)
print(snow_cover_km2_recalcule.shape)
```

```python
plt.plot(range(1, 13), snow_cover_km2_recalcule, marker="o")
plt.xlabel("Mois")
plt.ylabel("Superficie enneigée (km²)")
plt.title("Superficie moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

### plot avant / après regrid

```python
plt.plot(range(1, 13), snow_cover_km2_recalcule, marker="o", label="regrid")
plt.plot(range(1, 13), snow_cover_km2, marker="o", label="initial")
plt.xlabel("Mois")
plt.ylabel("Superficie enneigée (km²)")
plt.title("Superficie moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.legend()
plt.show()
```

```python
diff = snow_cover_km2 - snow_cover_km2_recalcule

plt.plot(range(1, 13), diff, marker="o")
plt.xlabel("Mois")
plt.ylabel("Diff (km2)")
plt.title("Difference superficie entre initial et regridded files")
plt.grid()
plt.show()
```
