```python
def get_data_2(path_snw, path_areacella, path_sftgif, path_sftlf):
    ds_snw = xr.open_dataset(os.path.expanduser(path_snw))
    ds_area = xr.open_dataset(os.path.expanduser(path_areacella))
    ds_sftgif = xr.open_dataset(os.path.expanduser(path_sftgif))
    ds_sftlf = xr.open_dataset(os.path.expanduser(path_sftlf))
    return ds_snw["snw"], ds_area["cell_area"], ds_sftgif["sftgif"], ds_sftlf["sftlf"]

def round_coords(*datasets):
    for ds in datasets:
        ds.coords['lat'] = ds.coords['lat'].round(4)
        ds.coords['lon'] = ds.coords['lon'].round(4)
    return datasets

def reindex_datasets(snw, cell_area, sftgif, sftlf):
    snw = snw.reindex_like(sftgif, method='nearest')
    cell_area = cell_area.reindex_like(sftlf, method='nearest')
    return snw, cell_area

def snowmass_sans_glaciers(snw, cell_area, sftlf, sftgif, start_year="1995", end_year="2014", lat_min=20):
    snw_masked = snw * (sftlf/100) * ((100 - sftgif)/100)
    if lat_min is not None:
        snw_masked = snw_masked.sel(lat=snw_masked.lat>=lat_min)
    if sftlf is not None:
        snw_masked = snw_masked.where(sftlf>0)
    snw_masked = snw_masked.sel(time=slice(start_year, end_year))
    snowmass_EC = snw_masked.groupby("time.month").mean(dim="time")
    snowmass_EC = (snowmass_EC * cell_area).sum(dim=["lat", "lon"], skipna=True)
    #snw_masked = snw_masked.where(snw_masked <=1, snw_masked / 100)
    #snw_masked = snc_masked.clip(0, 1)
    snowmass_gt_EC = snowmass_EC / 1e12
    return snw_masked, snowmass_gt_EC

def plot_monthly_mean(snowmass_gt_EC, xlabel="month", ylabel="Snowmass (Gt)", title="Quantité totale mensuelle de neige"):
    mois = snowmass_gt_EC["month"].values  # récupère les noms des mois
    valeurs = snowmass_gt_EC.values
    plt.plot(mois, valeurs, marker="o")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.title(title)
    plt.grid()
    plt.show()

def plot_spatial_map(snw_masked, time_index=0, title="Carte couverture neigeuse"):
    plt.figure(figsize=(10, 5))
    ax = plt.axes(projection=ccrs.PlateCarree())
    snw_masked.isel(time=time_index).plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(), cmap='viridis')
    ax.coastlines()
    plt.title(title)
    plt.show() 
```

```python
snw, area, sftgif, sftlf = get_data_2(
	"~/private-storage/output_data/interp_snowmass/EC-Earth_snowmass_reprojete.nc",
	"~/private-storage/areacella_recalcule_cdo_bis/EC-Earth3_gridarea_hist_recalcule.nc",
	"~/private-storage/interp_sftgif_bis/EC-Earth3_sftgif_reprojete.nc",
	"~/private-storage/interp_sftlf_bis/EC-Earth3_sftlf_reprojete.nc"
)
```

```python
mois = ["janvier", "février", "mars", "avril", "mai", "juin",
        "juillet", "août", "septembre", "octobre", "novembre", "décembre"]

snowmass_gt_EC["month"] = mois
```

```python
def get_data(path_snw):
    ds_snw = xr.open_dataset(path_snw)
    #ds_area = xr.open_dataset(path_areacella)
    #ds_sftlf = xr.open_dataset(path_sftlf)
    return ds_snw

def extraction_variable_and_monthly_mean(ds_snw, variable="snw", start_year="1995", end_year="2014", lat_min=20):
    snw = ds_snw[variable].sel(time=slice(start_year, end_year))

    if lat_min is not None:
        snw = snw.sel(lat=snw.lat >= lat_min)

    snw = snw.where(snw < 1e19, float("nan"))

    #max_val = snw.max().item()
    #if max_val > 1.5:
        #print(f"La variable '{variable}' est en %. Conversion en fraction.")
        #snw = snw / 100

    #if sftlf is not None:
        #snw=snw.where(sftlf>0)

    #snw = snw.clip(0, 1)

    snowmass = snw.groupby("time.month").mean(dim="time")
    monthly_sum = snowmass.cumsum(dim=("lat", "lon"))

    return snw, snowmass, monthly_sum
```
