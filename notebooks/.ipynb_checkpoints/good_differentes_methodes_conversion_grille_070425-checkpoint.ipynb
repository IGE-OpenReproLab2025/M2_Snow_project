```python
snow_cover_km2_2 = (snow_cover_2 * (cell_area_ini_2 * 1e-6) ).groupby(snow_cover_2["month"]).sum(dim=["lat", "lon"], skipna=True)

print(snow_cover_km2_2)
print(snow_cover_km2_2.shape)
```

```python
plt.plot(range(1, 13), snow_cover_km2, marker="o")
plt.xlabel("Mois")
plt.ylabel("Superficie enneigÃ©e (kmÂ²)")
plt.title("Superficie moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

```python
import xarray as xr
import os
import subprocess

# --- 1. ParamÃ¨tres et chemins ---
# Fichier neige
snc_path = "/home/jovyan/private-storage/historical-LImon-snc/snc_LImon_BCC-CSM2-MR_historical_r1i1p1f1_gn_185001-201412.nc"

# Masque terrestre
mask_path = "/home/jovyan/private-storage/masques/masques/pseudo_sftlf/sftlf_BCC-CSM2-MR.nc"  # Ã  adapter si besoin

# Grille cible (fichier NetCDF)
grid_target = "/home/jovyan/private-storage/grid/nhsce_mon.v01r01_19661004_20250303_1.0x1.0.nc"

# Dossier et nom de sortie
output_path = "/home/jovyan/private-storage/interp_models_hist/snc_interp_masked.nc"

# Fichier temporaire
temp_masked_file = "snc_masked.nc"

# --- 2. Charger les donnÃ©es ---
snc = xr.open_dataset(snc_path)["snc"]
mask = xr.open_dataset(mask_path)["sftlf"]

# --- 3. Appliquer le masque (1 = terre, 0 = ocÃ©an) ---
mask_bool = xr.where(mask > 0, 1, 0)

# Assurer les mÃªmes dimensions de lat/lon
if not snc.lat.equals(mask.lat) or not snc.lon.equals(mask.lon):
    print("â ï¸ Le masque n'est pas sur la mÃªme grille que la neige. Reprojection nÃ©cessaire.")
    mask_bool = mask_bool.interp_like(snc, method="nearest")  # ou conservative si besoin

# --- 4. Masquer la neige sur ocÃ©an ---
snc_masked = snc.where(mask_bool == 1)

# --- 5. Sauvegarder temporairement ---
snc_masked.to_netcdf(temp_masked_file)
print(f"â Fichier temporaire sauvegardÃ© : {temp_masked_file}")

# --- 6. Interpolation avec CDO ---
cmd = [
    "cdo",
    f"remapcon,{grid_target}",
    temp_masked_file,
    output_path
]

try:
    print(f"ð ExÃ©cution de la commande CDO : {' '.join(cmd)}")
    subprocess.run(cmd, check=True)
    print(f"â Fichier interpolÃ© sauvegardÃ© ici : {output_path}")
except subprocess.CalledProcessError as e:
    print("â Erreur pendant l'interpolation CDO :", e)

# --- 7. Nettoyage optionnel ---
# os.remove(temp_masked_file)
```
