### Variable areacella

```python
path_areacella_BCC_CSM2_initial = os.path.join("/home/jovyan/private-storage/masques/masques/pseudo_areacella/", "areacella_BCC-CSM2-MR.nc")

areacella_BCC_CSM2_initial = xr.open_dataset(path_areacella_BCC_CSM2_initial)
```

```python
cell_area_ini = areacella_BCC_CSM2_initial["areacella"]
```

### Arrondi à 4 ch

```python
snow_cover_ini.coords['lat'] = snow_cover_ini.coords['lat'].round(4)
snow_cover_ini.coords['lon'] = snow_cover_ini.coords['lon'].round(4)

cell_area_ini.coords['lat'] = cell_area_ini.coords['lat'].round(4)
cell_area_ini.coords['lon'] = cell_area_ini.coords['lon'].round(4)

snow_cover_ini = snow_cover_ini.reindex_like(sftgif, method='nearest')
cell_area_ini = cell_area_ini.reindex_like(sftlf, method='nearest')

sftgif.coords['lat'] = sftgif.coords['lat'].round(4)
sftgif.coords['lon'] = sftgif.coords['lon'].round(4)

sftlf.coords['lat'] = sftlf.coords['lat'].round(4)
sftlf.coords['lon'] = sftlf.coords['lon'].round(4)
```

### Calculs

```python
snc = snow_cover_ini * cell_area_ini *1e-6
```

```python
sna1 = snc * (sftlf/100)
```

```python
sna2 = sna1 * ((100 - sftgif)/100)
sna2
```

```python
plt.figure(figsize=(10, 5))
sna2.isel(time=1799).plot()
```

```python
sna2_ref = sna2.sel(time=slice("1995", "2014"))
sna2_ref
```

```python
snow_cover = sna2.groupby("time.month").mean(dim="time")
snow_cover
```

```python
monthly_mean = snow_cover.sum(dim=("lat", "lon"))

plt.plot(range(1, 13), monthly_mean, marker = "o")
plt.xlabel("Mois")
plt.ylabel("Couverture neigeuse moyenne")
plt.title("Moyenne mensuelle de la couverture neigeuse")
plt.grid()
plt.show()
```

## En fonctions

```python
def get_data_2(path_snc, path_areacella, path_sftgif, path_sftlf):
    ds_snc = xr.open_dataset(path_snc)
    ds_area = xr.open_dataset(path_areacella)
    ds_sftgif = xr.open_dataset(path_sftgif)
    ds_sftlf = xr.open_dataset(path_sftlf)
    return ds_snc, ds_area, ds_sftgif, ds_sftlf

def superficie_sans_glaciers(snc, areacella, sftlf, sftgif, start_year="1995", end_year="2014"):
    area_km2 = areacella * 1e-6
    snc_km2 = (snc * area_km2) * (sftlf/100) * ((100 - sftgif)/100)
    snc_km2 = snc_km2.sel(time=slice(start_year, end_year))
    return snc_km2

def monthly_mean_calculation(snc_km2):
    monthly_mean = (snc_km2.groupby("time.month").mean(dim="time")).sum(dim=("lat", "lon"))

def plot_monthly_mean(snc_km2, xlabel="Mois", ylabel="Superficie (km2)", title="Moyenne mensuelle de la couverture neigeuse"):
    plt.plot(range(1, 13), snc_km2, marker="o")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid()
    plt.show()
```
