```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import os
```

```python
%watermark --iversions
```

```python
def get_data_2(path_snc, path_areacella, path_sftgif, path_sftlf):
    ds_snc = xr.open_dataset(os.path.expanduser(path_snc))
    ds_area = xr.open_dataset(os.path.expanduser(path_areacella))
    ds_sftgif = xr.open_dataset(os.path.expanduser(path_sftgif))
    ds_sftlf = xr.open_dataset(os.path.expanduser(path_sftlf))
    return ds_snc["snc"], ds_area["cell_area"], ds_sftgif["sftgif"], ds_sftlf["sftlf"]

def round_coords(*datasets):
    for ds in datasets:
        ds.coords['lat'] = ds.coords['lat'].round(4)
        ds.coords['lon'] = ds.coords['lon'].round(4)
    return datasets

def reindex_datasets(snc, cell_area, sftgif, sftlf):
    snc = snc.reindex_like(sftgif, method='nearest')
    cell_area = cell_area.reindex_like(sftlf, method='nearest')
    return snc, cell_area

def superficie_sans_glaciers(snc, cell_area, sftlf, sftgif, start_year="1995", end_year="2014", lat_min=20):
    snc_masked = snc * (sftlf/100) * ((100 - sftgif)/100)
    if lat_min is not None:
        snc_masked = snc_masked.sel(lat=snc_masked.lat>=lat_min)
    snc_masked = snc_masked.sel(time=slice(start_year, end_year))
    snc_masked = snc_masked.where(snc_masked <=1, snc_masked / 100)
    snc_masked = snc_masked.clip(0, 1)
    snc_km2 = snc_masked * cell_area * 1e-6
    return snc_masked, snc_km2

def monthly_mean_calculation(snc_km2):
    return snc_km2.groupby("time.month").mean(dim="time").sum(dim=("lat", "lon"))

def plot_monthly_mean(snc_monthly_mean, xlabel="Mois", ylabel="Superficie (km2)", title="Moyenne mensuelle de la couverture neigeuse"):
    plt.plot(range(1, 13), snc_monthly_mean, marker="o")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid()
    plt.show()

def plot_spatial_map(snc_masked, time_index=0, title="Carte couverture neigeuse"):
    plt.figure(figsize=(10, 5))
    ax = plt.axes(projection=ccrs.PlateCarree())
    snc_masked.isel(time=time_index).plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(), cmap='viridis')
    ax.coastlines()
    plt.title(title)
    plt.show() 
```

```python
snc, area, sftgif, sftlf = get_data_2(
	"~/private-storage/interp_models_hist_bis/IPSL-CM6A-LR_hist_reprojete.nc",
	"~/private-storage/areacella_recalcule_cdo_bis/IPSL-CM6A-LR_gridarea_hist_recalcule.nc",
	"~/private-storage/interp_sftgif_bis/IPSL-CM6A-LR_sftgif_reprojete.nc",
	"~/private-storage/interp_sftlf_bis/IPSL-CM6A-LR_sftlf_reprojete.nc"
)
```

```python
snc, area, sftgif, sftlf = round_coords(snc, area, sftgif, sftlf)
snc, area = reindex_datasets(snc, area, sftgif, sftlf)
```

```python
snc_masked, snc_km2 = superficie_sans_glaciers(snc, area, sftlf, sftgif)
```

```python
plot_spatial_map(snc_masked, title="Superficie neigeuse sans glaciers")
```

```python
monthly = monthly_mean_calculation(snc_km2)
```

```python
plot_monthly_mean(monthly)
```
