# Plot des changement de snow cover par rapport à référence selon GSAT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

## à 50% de neige

```python
# méthode avec fichiers historiques
winter_snow_change = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - snow_extent_change_winter.csv', sep=',')
winter_snow_change = winter_snow_change.drop(columns=['Unnamed: 0'])
winter_snow_change
```

```python
spring_snow_change = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - snow_extent_change_spring.csv', sep=',')
spring_snow_change = spring_snow_change.drop(columns=['Unnamed: 0'])
spring_snow_change
```

```python
summer_snow_change = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - snow_extent_change_summer.csv', sep=',')
summer_snow_change = summer_snow_change.drop(columns=['Unnamed: 0'])
summer_snow_change
```

```python
autumn_snow_change = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - snow_extent_change_autumn.csv', sep=',')
autumn_snow_change = autumn_snow_change.drop(columns=['Unnamed: 0'])
autumn_snow_change
```

```python
# Conversion des virgules en points et des strings en float
winter_snow_change["GWL"] = winter_snow_change["GWL"].str.replace(',', '.').astype(float)
winter_snow_change["snow_change"] = winter_snow_change["snow_change"].str.replace(',', '.').astype(float)

# Tracé
plt.figure(figsize=(8, 6))
sns.regplot(data=winter_snow_change, x="GWL", y="snow_change", ci=None, color="blue", line_kws={"color": "red"})

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (1995-2014 period)")
plt.ylabel("Snow Extent Change (%)")
plt.title("Relation entre changement d'enneigement et GWL\n(Winter)")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Conversion des virgules en points et des strings en float
spring_snow_change["GWL"] = spring_snow_change["GWL"].str.replace(',', '.').astype(float)
spring_snow_change["snow_change"] = spring_snow_change["snow_change"].str.replace(',', '.').astype(float)

# Tracé
plt.figure(figsize=(8, 6))
sns.regplot(data=spring_snow_change, x="GWL", y="snow_change", ci=None, color="blue", line_kws={"color": "red"})

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (relative to 1995-2014)")
plt.ylabel("Snow Cover Extent Change (%)")
plt.title("Spring")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Conversion des virgules en points et des strings en float
summer_snow_change["GWL"] = summer_snow_change["GWL"].str.replace(',', '.').astype(float)
summer_snow_change["snow_change"] = summer_snow_change["snow_change"].str.replace(',', '.').astype(float)

# Tracé
plt.figure(figsize=(8, 6))
sns.regplot(data=summer_snow_change, x="GWL", y="snow_change", ci=None, color="blue", line_kws={"color": "red"})

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (relative to 1995-2014)")
plt.ylabel("Snow Cover Extent Change (%)")
plt.title("Summer")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Conversion des virgules en points et des strings en float
autumn_snow_change["GWL"] = autumn_snow_change["GWL"].str.replace(',', '.').astype(float)
autumn_snow_change["snow_change"] = autumn_snow_change["snow_change"].str.replace(',', '.').astype(float)

# Tracé
plt.figure(figsize=(8, 6))
sns.regplot(data=autumn_snow_change, x="GWL", y="snow_change", ci=None, color="blue", line_kws={"color": "red"})

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (relative to 1995-2014)")
plt.ylabel("Snow Cover Extent Change (%)")
plt.title("Autumn")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Tracé
plt.figure(figsize=(8, 6))
sns.regplot(data=winter_snow_change, x="GWL", y="snow_change", ci=None, color="dodgerblue", line_kws={"color": "dodgerblue", "linestyle": "--"}, label ="winter")
sns.regplot(data=spring_snow_change, x="GWL", y="snow_change", ci=None, color="mediumseagreen", line_kws={"color": "mediumseagreen", "linestyle": "--"}, label = "spring")
sns.regplot(data=summer_snow_change, x="GWL", y="snow_change", ci=None, color="goldenrod", line_kws={"color": "goldenrod", "linestyle": "--"}, label = "summer")
sns.regplot(data=autumn_snow_change, x="GWL", y="snow_change", ci=None, color="sienna", line_kws={"color": "sienna", "linestyle": "--"}, label = "autumn")

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (relative to 1995-2014)")
plt.ylabel("Snow Cover Extent Change (%)")
plt.legend(loc="lower left")
#plt.title("Autumn")

plt.grid(True)
plt.tight_layout()
#plt.savefig("snow_cover_extent_change.png")
plt.show()
```

## à 15% de neige

```python
# méthode avec fichiers historiques
winter_snow_change_15 = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - change_winter.csv', sep=',')
winter_snow_change_15 = winter_snow_change_15.drop(columns=['Unnamed: 0'])
```

```python
spring_snow_change_15 = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - change_spring.csv', sep=',')
spring_snow_change_15 = spring_snow_change_15.drop(columns=['Unnamed: 0'])
```

```python
autumn_snow_change_15 = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - change_autumn.csv', sep=',')
autumn_snow_change_15 = autumn_snow_change_15.drop(columns=['Unnamed: 0'])
```

```python
summer_snow_change_15 = pd.read_csv('~/private-storage/fichiers_csv/Comparaison entre modèles - change_summer.csv', sep=',')
summer_snow_change_15 = summer_snow_change_15.drop(columns=['Unnamed: 0'])
```

```python
winter_snow_change_15["GWL"] = winter_snow_change_15["GWL"].str.replace(',', '.').astype(float)
winter_snow_change_15["snow_change"] = winter_snow_change_15["snow_change"].str.replace(',', '.').astype(float)
```

```python
spring_snow_change_15["GWL"] = spring_snow_change_15["GWL"].str.replace(',', '.').astype(float)
spring_snow_change_15["snow_change"] = spring_snow_change_15["snow_change"].str.replace(',', '.').astype(float)

autumn_snow_change_15["GWL"] = autumn_snow_change_15["GWL"].str.replace(',', '.').astype(float)
autumn_snow_change_15["snow_change"] = autumn_snow_change_15["snow_change"].str.replace(',', '.').astype(float)

summer_snow_change_15["GWL"] = summer_snow_change_15["GWL"].str.replace(',', '.').astype(float)
summer_snow_change_15["snow_change"] = summer_snow_change_15["snow_change"].str.replace(',', '.').astype(float)
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(
    data=winter_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="winter", marker='o', scatter_kws={"s": 60}
)  # cercle
sns.regplot(
    data=spring_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="spring", marker='s', scatter_kws={"s": 60}
)  # carré
sns.regplot(
    data=summer_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="summer", marker='^', scatter_kws={"s": 60}
)  # triangle
sns.regplot(
    data=autumn_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="autumn", marker='*', scatter_kws={"s": 100}
)  # étoile

plt.axhline(0, color='gray', linestyle='--')  # ligne horizontale à y = 0
plt.axvline(0, color='gray', linestyle='--')  # ligne verticale à x = 0

plt.xlabel("GWL (relative to 1995-2014)", fontsize='14')
plt.ylabel("Snow Cover Extent Change (%)", fontsize='14')
plt.legend(loc="lower left", fontsize='13')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.title("Autumn")

plt.grid(True)
plt.tight_layout()
#plt.savefig("figures/snow_cover_extent_change_15%.png")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Couleurs sobres, contrastées et sans bleu/rouge/orange
colors = {
    "winter": "cornflowerblue",  
    "spring": "green", 
    "summer": "mediumorchid", 
    "autumn": "peru", 
}

sns.regplot(
    data=winter_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="winter", marker='o', scatter_kws={"s": 60}, color=colors["winter"]
)

sns.regplot(
    data=spring_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="spring", marker='s', scatter_kws={"s": 60}, color=colors["spring"]
)

sns.regplot(
    data=summer_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="summer", marker='^', scatter_kws={"s": 60}, color=colors["summer"]
)

sns.regplot(
    data=autumn_snow_change_15, x="GWL", y="snow_change", ci=None,
    label="autumn", marker='*', scatter_kws={"s": 100}, color=colors["autumn"]
)

plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')

plt.xlabel("GWL (relative to 1995-2014)", fontsize=14)
plt.ylabel("Snow Cover Extent Change (%)", fontsize=14)
plt.legend(loc="lower left", fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.grid(True)
plt.tight_layout()
plt.savefig("figures/snow_cover_extent_change_15%_bis.png")
plt.show()
```
